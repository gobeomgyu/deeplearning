```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

# 1. 데이터 로딩
df = pd.read_csv("C:\\Users\\LG\\Downloads\\titanic.csv", index_col=0)
df

# 2. 결측치 확인
df.isnull().sum()

# 3. 결측치 제거(평균값으로 대체)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.dropna(inplace=True)
df.isnull().sum()

# 4.레이블 분포 확인
print(df['Survived'].value_counts())

# 비율 확인
print(df['Survived'].value_counts(normalize=True))

# Survived 컬럼의 클래스 분포 시각화
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df, palette='viridis')
plt.title('Survived Class Distribution by Sex')
plt.show()

#5. 여러 개의 불필요한 컬럼 제거
df = df.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1) #axis=1 -> 열을 기준으로 제거
df

df['Survived'].value_counts()

# 6. Sex 컬럼 인코딩 (Label Encoding)
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])  # 'male'은 0, 'female'은 1로 변환

df['Survived'].value_counts()
df['Sex'].value_counts()
```
